```python
# Nama     : Hylmi Farras Madjid
# Tanggal  : 23/09/2023
# Deskripsi: GDSC Project

import pandas as pd
import matplotlib.pyplot as plt
dat = pd.read_csv("Dataset_Project_GDSC.csv") #Definisikan dataframe as df len(df.columns)
len(dat) #hitung jumlah baris
```

```python
maxi = dat["produktivitas_jagung"].idxmax() #Kabupaten dengan produktivitas tanaman jagung paling banyak
dat[maxi:maxi+1]

```

```python
mini = dat["produktivitas_jagung"].idxmin() #Kabupaten dengan produktivitas tanaman jagung paling sedikit
dat[mini:mini+1]
```

```python
#mengambil data dengan produktivitas tanaman jagung secara keseluruhan paling banyak
dat.max()
```

```python
#mencari nilai Rata – rata, Standar Deviasi, nilai minimum, nilai maksimal, persentase 25%, persentase 50%, dan persentase 75%
dat.describe()
```

```python
#mengurutkan tahun yang terdapat di dataset ini
dat.sort_values(["tahun"], ascending = True).groupby("tahun").head(1)
```

```python
#mengurutkan data produktivitas tanaman jagung dari yang terkecil hingga yang terbesar
dat.sort_values(["produktivitas_jagung"], ascending = True)
```

```python
dat["tahun"].corr(dat["produktivitas_jagung"])
```

```python
diagram= dat[["nama_kabupaten_kota", "produktivitas_jagung"]]
diagram= diagram.groupby('nama_kabupaten_kota').sum()
diagram.plot.pie(subplots=True, title="Persentase Produktivitas Tanaman Jagung per Kabupaten/Kota", autopct= '%1.0f%%',startangle=90, shadow=False,legend = False,fontsize=14, figsize=(10, 10) )
```

```python
pie = dat["tahun"].value_counts() 
pie.plot(kind = "pie")
plt.show()
```
